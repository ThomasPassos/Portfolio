# Análise de dados sobre a venda de jogos

```python
import numpy as np
import pandas as pd
import seaborn.objects as so
```

## 1 - Preparação

```python
df = pd.read_csv("game_sales.csv")
display(df)
```

```python
df.info()
```

```python
df.describe()
```

```python
(so.Plot(df, y="Global_Sales", x="Critic_Score", color="Genre")
    .facet("Genre", wrap=4)
    .add(so.Dots(), so.Jitter(0.3)).layout(size=(10, 10)))
```

```python
(
    so.Plot(df, x="Critic_Score")
     .facet("Genre", wrap=5)
     .add(so.Bars(), so.Hist())
     .label(x="Critic_Score")
     .layout(size=(10, 15))
)
```
